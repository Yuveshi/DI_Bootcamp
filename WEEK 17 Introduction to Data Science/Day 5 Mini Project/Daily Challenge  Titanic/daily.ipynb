```python
import pandas as pd       # For data manipulation
import numpy as np        # For numerical operations
import matplotlib.pyplot as plt  # For visualization
```

```python
# Load the Titanic dataset from a CSV file
df = pd.read_csv('titanic.csv')

# Preview the first few rows of the dataset
print(df.head())
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())

```

```python
# Count the number of male and female passengers
gender_counts = df['Sex'].value_counts()

# Plot a pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'pink'])
plt.title('Male vs Female Passengers')
plt.show()
```

```python
# Plot a histogram for the 'Age' column
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Count the number of survivors and non-survivors
survival_counts = df['Survived'].value_counts()

# Plot a bar chart
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.xticks([0, 1], ['Did not survive', 'Survived'])
plt.title('Survival Count')
plt.xlabel('Survival')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Create a grouped DataFrame by 'Sex' and 'Survived'
gender_survival = df.groupby(['Sex', 'Survived']).size().unstack()

# Plot a stacked bar chart
gender_survival.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Plot a scatter plot to show the correlation between fare and age
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title('Fare vs Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Fill missing 'Age' values with the mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the 'Fare' column
df['Fare'] = scaler.fit_transform(df[['Fare']])
```

```python
# Convert 'Sex' column to numerical values using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Sex'], drop_first=True)

# Display the first few rows to verify the transformation
print(df_encoded.head())
```

```python
# Calculate the survival rate for different genders
gender_survival_rate = df.groupby('Sex')['Survived'].mean()
print(gender_survival_rate)
```

```python
# Calculate the survival rate for different passenger classes (Pclass)
class_survival_rate = df.groupby('Pclass')['Survived'].mean()
print(class_survival_rate)
```

```python
# Find the passenger with the highest fare
highest_fare_passenger = df[df['Fare'] == df['Fare'].max()]
print(highest_fare_passenger)
```
