```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100)
```

```python
y = x**2
```

```python
plt.plot(x, y)
plt.title('Square of Numbers')
plt.xlabel('X values')
plt.ylabel('Y values (X squared)')
plt.xlim(0, 20)
plt.ylim(0, 400)
plt.savefig('plot_task1.png')
plt.show()
```
