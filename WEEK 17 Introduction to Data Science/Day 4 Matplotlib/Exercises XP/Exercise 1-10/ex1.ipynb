```python
import numpy as np
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, color='tab:blue', label='sin(x)')  # Line plot for sin(x)
plt.scatter(x, y2, color='red', label='cos(x)', marker='o')  # Scatter plot for cos(x)
plt.legend()  # Add a legend
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2)  # Creates 2 subplots in a single figure

ax1.plot(x, y1, color='tab:blue', label='sin(x)')
ax1.set_title("Sine Function")
ax1.legend()

ax2.scatter(x, y2, color='red', label='cos(x)', marker='o')
ax2.set_title("Cosine Function")
ax2.legend()

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
